```
#ACCPECT EMPLOYEE BASIC SALARY AND FIND THE GROSS SALARY?
BASIC = int(input("ENTER BASIC SALARY TO FIND GROSS SALARY:"))
if BASIC <= 8000:
   print() # HRA IS 60% ON BASIC, DA IS 40% ON BASIC
   HRA = 0.60 * BASIC
   DA = 0.40 * BASIC
elif BASIC <= 12000:
   print() # HRA IS 50% ON BASIC, DA IS 20% ON BASIC
   HRA = 0.50 * BASIC
   DA = 0.20* BASIC
else:
   print() # HRA IS 40% ON BASIC, DA IS 100% ON BASIC
   HRA = 0.40 * BASIC
   DA = 0.100 * BASIC
# Calculating Gross Salary
GROSS_SALARY = BASIC + HRA + DA
print("BASIC SALARY:", BASIC)
print("HRA:", HRA)
print("DA:", DA)
print("GROSS SALARY:", GROSS_SALARY)
```
